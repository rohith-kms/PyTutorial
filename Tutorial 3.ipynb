# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

```python
# linspace

xs = np.linspace(-10,10,1000)
```

```python
# Plotting

ys = erf(xs)
```

```python
plt.plot(xs,ys)
```

```python
plt.plot(xs,ys)
plt.plot(xs,ys*1.2)
plt.plot(xs,ys*0.8)
```

```python
# basic formatting of plots

# recommended syntax for plotting

fig, ax = plt.subplots(figsize=(10,8))

ax.plot(xs,ys, linestyle='--')
ax.plot(xs+1,ys*1.2, linestyle=':', color='black')
ax.plot(xs,ys*0.8, color='#FFFF8F')

ax.set_xlabel('Label 1')
ax.set_ylabel('Label 2')
ax.legend(['c1','c2','c3'])
ax.set_facecolor('#BDC3C7')
```

```python
fig, ax = plt.subplots(nrows=1,ncols=3)

ax[0].plot(xs,ys, linestyle='--')
ax[1].plot(xs+1,ys*1.2, linestyle=':', color='black')
ax[2].plot(xs,ys*0.8, color='#FFFF8F')

for i in range(3):
    
    ax[i].set_xlabel('label 1')
```

```python
# function z = f(x,y)

def z(x,y):
    
    return erf(x)+np.sin(y)

ys = np.linspace(-10,10,1000)
```

```python
np.shape(xs)
```

```python
np.shape(ys)
```

```python
grid = np.meshgrid(xs,ys)

zs = z(grid[0],grid[1])

np.shape(zs)
```

```python
%matplotlib notebook

fig = plt.figure()
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(xs, ys, zs)
```
